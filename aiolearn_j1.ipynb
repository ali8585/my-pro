```python
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("/content/photo10955484635.jpg")
plt.imshow(image)
plt.show()
```

```python
model = cv2.CascadeClassifier("/content/model.xml")
```

```python
face = model.detectMultiScale(image)
```

```python
x = face[0][0]
y = face[0][1]
a = face[0][2]
b = face[0][3]
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.rectangle(image, (x, y) , (x+a, y+b) , (255, 0, 0) , 4)
plt.imshow(image)
plt.show()
```
